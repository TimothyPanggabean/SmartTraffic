```python
import os
import xml.etree.ElementTree as ET
```

```python
def convert_box(size, box):

  dw, dh = 1. / size[0], 1. / size[1]
  x, y, w, h = (box[0] + box[1]) / 2.0 - 1, (box[2] + box[3]) / 2.0 - 1, box[1] - box[0], box[3] - box[2]
  return x * dw, y * dh, w * dw, h * dh
```

```python
yolo_ambulance = os.path.join('ambulance','yolo_anotasi')
yolo_pemadam = os.path.join('Pemadam','yolo_anotasi')

xml_ambulance = os.path.join('ambulance','anotasi')
xml_pemadam = os.path.join('Pemadam','Anotasi')

if not os.path.exists(yolo_ambulance):
  os.mkdir(yolo_ambulance)

if not os.path.exists(yolo_pemadam):
  os.mkdir(yolo_pemadam)
```

Label: Car=0, Motorcycle=1, Ambulance=2, Pemadam=3

```python
for filename in os.listdir('ambulance/anotasi'):
  tree = ET.parse(os.path.join(xml_ambulance, filename))
  root = tree.getroot()
  for size in root.findall('size'):
    w = int(size.find('width').text)
    h = int(size.find('height').text)
  for object_tag in root.findall('object'):
    xmin = int(object_tag.find('bndbox').find('xmin').text)
    xmax = int(object_tag.find('bndbox').find('xmax').text)
    ymin = int(object_tag.find('bndbox').find('ymin').text)
    ymax = int(object_tag.find('bndbox').find('ymax').text)

  bbox = convert_box((w, h), [xmin, xmax, ymin, ymax])

  txt_filename = (os.path.splitext(filename)[0]+'.txt')
  with open(os.path.join(yolo_ambulance, txt_filename), 'w') as f:
    f.write('2 {} {} {} {}'.format(bbox[0], bbox[1], bbox[2], bbox[3]))
```

```python
for filename in os.listdir('Pemadam/Anotasi'):
  tree = ET.parse(os.path.join(xml_pemadam, filename))
  root = tree.getroot()
  for size in root.findall('size'):
    w = int(size.find('width').text)
    h = int(size.find('height').text)
  for object_tag in root.findall('object'):
    xmin = int(object_tag.find('bndbox').find('xmin').text)
    xmax = int(object_tag.find('bndbox').find('xmax').text)
    ymin = int(object_tag.find('bndbox').find('ymin').text)
    ymax = int(object_tag.find('bndbox').find('ymax').text)

    bbox = convert_box((w, h), [xmin, xmax, ymin, ymax])

    txt_filename = (os.path.splitext(filename)[0]+'.txt')
    with open(os.path.join(yolo_pemadam, txt_filename), 'w') as f:
      f.write('3 {} {} {} {}'.format(bbox[0], bbox[1], bbox[2], bbox[3]))
```
